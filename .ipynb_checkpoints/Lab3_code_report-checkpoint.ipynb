## 2) Predicting Diabetes on the Pima dataset


### 2.1 Know the data


```python
# Create scatterplots between pairs of variables with class coloring
df['Class'] = y1
sns.pairplot(df, hue='Class', palette='tab10')
plt.show()
```

**What are the variables that seem to help predict the class? Do those variable make sense from a
medical perspective ?**

Seaborn can help us doing a better vizualisation plot than matpllotlib for this case.

By using a pairplot to visualize the relationships between variables, we aimed to identify key factors for predicting diabetes. Real class distinctions (diabetes vs. non-diabetes) were represented by color. Notably, the variable **glucose** exhibited a clear separation between the two classes, indicating its significance in predicting diabetes. On the diagonal, histograms revealed the variable distributions. From this analysis, we can infer:

- **Glucose**: Elevated levels are associated with diabetes.
- **Insulin** and **BMI**: These also play a role but with less distinct separation than **glucose**
- **Skin thickness**: Shows significant separation.
- **Other variables**: Show little to no noticeable influence as they are not well separated.

In a medical context, glucose is a primary diabetes indicator, confirmed by healthcare professionals who rely on fasting glucose levels for diagnosis. Diabetic individuals typically have hyperglycemia. Consequently, our dataset's emphasis on glucose as a predictor aligns with medical knowledge. While insulin and BMI are secondary factors, insulin lowers blood sugar, and some individuals with diabetes have higher BMIs.

```python
from sklearn.model_selection import train_test_split

# Define the number of samples to keep for training/validation
n_train_validation = 300

# Split the data into training/validation and test sets
x_train_val, x_test, y_train_val, y_test = train_test_split(
    df.drop('Class', axis=1), df['Class'], test_size=len(df) - n_train_validation, random_state=42)

# Print the shapes of the resulting sets
print("Training/Validation Set (X):", x_train_val.shape)
print("Training/Validation Set (Y):", y_train_val.shape)
print("\nTest Set (X):", x_test.shape)
print("Test Set (Y):", y_test.shape)
```

Now we split the data between train and test sets as it is not done yet like for the digits dataset. For this lab, I'm using the function **train_test_split** from sklearn, it makes the code cleaner but i still think that doing it manually remains more visual in the code.

```python
# Calculate mean and variance of the original data
mean_before_scaling = x_train_val.mean()
variance_before_scaling = x_train_val.var()
```

```python
# Print the statistics before and after scaling
print("Original Data - Mean:")
print(mean_before_scaling)
print("Original Data - Variance:")
print(variance_before_scaling)
```

As we saw before, the means and variance of the features are nothing alike. We're gonna scale them and see the difference on these 2 informations.

```python
from sklearn.preprocessing import StandardScaler

# Instantiate the StandardScaler
scaler = StandardScaler()

# Fit and transform the training/validation data
x_train_val_scaled = scaler.fit_transform(x_train_val)

# Transform the test data using the same scaler
x_test_scaled = scaler.transform(x_test)

# Check the variances of the scaled data
print("Variances of scaled training/validation data:")
print(np.var(x_train_val_scaled, axis=0))

print("Variances of scaled test data:")
print(np.var(x_test_scaled, axis=0))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))

# Plot for the training dataset
axes[0].boxplot(x_train_val_scaled, patch_artist=True, showmeans=True)
axes[0].set_title('Boxplots of the features for scaled training dataset')
axes[0].set_xticks(np.arange(1, 9))
axes[0].set_xticklabels(varnames, rotation=45)

# Plot for the testing dataset
axes[1].boxplot(x_test_scaled, patch_artist=True, showmeans=True)
axes[1].set_title('Boxplots of the features for scaled testing dataset')
axes[1].set_xticks(np.arange(1, 9))
axes[1].set_xticklabels(varnames, rotation=45)

plt.show()
```


The variances of the scaled training and validation data are all equal to 1, indicating that the features have been successfully standardized and have the same scale.

However, in the scaled test data, we observe slight variations in the variances of individual features, with values slightly above or below 1. This suggests that there may be some differences in feature distributions between the training and test datasets. It's important to note these differences when making predictions on the test data, as the feature scales may not be identical to the training data due to variations in the test dataset.

Let's see if these difference of the variances on the scaled testing data are signicative in the distributions of features.

```python
# Number of features
num_features = x_test.shape[1]

# Create a figure with subplots for all features
fig, axes = plt.subplots(num_features, 2, figsize=(12, 2 * num_features))

for feature_index in range(num_features):
    feature_name = varnames[feature_index]  # Get the feature name
    # Plot histograms of the feature before and after scaling
    axes[feature_index, 0].hist(x_test.values[:, feature_index], bins=30, color='b', alpha=0.7, label='Original Data')
    axes[feature_index, 0].set_title(f'Feature: {feature_name} - Before Scaling')
    axes[feature_index, 0].legend()

    axes[feature_index, 1].hist(x_test_scaled[:, feature_index], bins=30, color='r', alpha=0.7, label='Scaled Data')
    axes[feature_index, 1].set_title(f'Feature: {feature_name} - After Scaling')
    axes[feature_index, 1].legend()

plt.tight_layout()
plt.show()
```

We can confidently conclude that these differences are not significant, as it is challenging to discern any notable disparities in the distributions of the features before and after scaling.

Now we have proper datasets that are scaled, we can start to work deeper with them.

### 2.2 Bayesian decision and linear classification


```python
# Initialize an empty DataFrame
data_models = {
    "Model": [],
    "Accuracy Default": [],
    "AUC Default": [],
    "Accuracy Best": [],
    "AUC Best": [],
}

df_models = pd.DataFrame(data_models)

# Create a function to apply the desired style to the DataFrame
def apply_style_to_dataframe(dataframe):
    styled_df = dataframe.style.set_properties(**{'text-align': 'center'})
    styled_df.set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]}])
    return styled_df

# Function to add model data to the DataFrame
def add_model_data(model_name, accuracy_default, auc_default, accuracy_best, auc_best, dataframe):
    new_data = {
        "Model": model_name,
        "Accuracy Default": accuracy_default,
        "AUC Default": auc_default,
        "Accuracy Best": accuracy_best,
        "AUC Best": auc_best,
    }
    # Create a new DataFrame with the new data
    new_df = pd.DataFrame(new_data, index=[0])
    # Concatenate the new DataFrame with the existing DataFrame
    dataframe = pd.concat([dataframe, new_df], ignore_index=True)
    return dataframe

# Function to display or print the styled DataFrame
def display_styled_dataframe(styled_df):
    return styled_df
```

#### LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, roc_auc_score, make_scorer
```

```python
# Create an LDA classifier with default parameters
lda = LinearDiscriminantAnalysis()

# Fit the LDA classifier on the training data
lda.fit(x_train_val_scaled, y_train_val)

# Make predictions on the test data
y_pred = lda.predict(x_test_scaled)

# Calculate accuracy on the test data
accuracy_default_lda = accuracy_score(y_test, y_pred)

# Calculate AUC (Area Under the ROC Curve) on the test data
# First, get the probability estimates for class 1
y_prob = lda.predict_proba(x_test_scaled)[:, 1]

# Calculate the AUC
auc_default_lda = roc_auc_score(y_test, y_prob)
```

```python
print(f"Accuracy: {accuracy_default_lda:.5f}")
print(f"AUC: {auc_default_lda:.5f}")
```

Arround 79% of patients are well classified with default parameters of the LDA. It is a good result, let's see the results diference after finding the best parameters.

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Create an LDA classifier
lda = LinearDiscriminantAnalysis()

param_grid = {'solver': ['lsqr', 'eigen'],
              'shrinkage': np.linspace(0,1,100)}

# Create a GridSearchCV object
grid_search = GridSearchCV(lda, param_grid, scoring=make_scorer(roc_auc_score), verbose=1, n_jobs=-1)

# Fit the GridSearchCV to the training data
grid_search_fit = grid_search.fit(x_train_val_scaled, y_train_val)
```

```python
# Get the results of the grid search
print("Best parameters:", grid_search_fit.best_params_)
```

```python
# Get the best estimator and best parameter
best_lda = grid_search_fit.best_estimator_
best_params = grid_search_fit.best_params_

# Evaluate the best estimator on the test data
best_lda.fit(x_train_val_scaled, y_train_val)
y_pred = best_lda.predict(x_test_scaled)
y_prob_best_lda = best_lda.predict_proba(x_test_scaled)[:, 1]
```

```python
accuracy_best_lda = accuracy_score(y_test, y_pred)
auc_best_lda = roc_auc_score(y_test, y_prob_best_lda)
```

```python
#Print the results
df_models = add_model_data("LDA", accuracy_default_lda, auc_default_lda, accuracy_best_lda, auc_best_lda, df_models)
display_styled_dataframe(apply_style_to_dataframe(df_models))
```

The results are slightly better with optimal parameters but it do not lead to a huge difference.

Thus, this linear method seems to work well on our problem.

```python
# Plot a simple stem of the coefs with the names of the features
plt.figure(figsize=(10, 6))
plt.stem(best_lda.coef_.flatten())
plt.xticks(range(num_features), varnames, rotation=30)
plt.title("Coefficients of the best LDA classifier")
plt.show()
```

Using the LDA method, we can access the feature weights. As previously observed in the pairplot at the beginning, it's clear that **glucose** is the most significant feature in the dataset, strongly associated with diabetes. This relationship aligns with common knowledge.

Surprisingly, the weight assigned to **pregnancies** is substantial. This might be attributed to pregnant women's increased food intake to support their growing child, along with potential changes in eating behaviors, sometimes leading to the consumption of high-fat and sweet foods.

BMI have a larger role in the prediction than I tough compared to insulin when I first interpret the pairplot. It thus very important too refer to several tools before taking any conclusions. Age is also a factor non negligable, which well known as risk for diabate increase with age.



#### QDA

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
# Create a QDA classifier with default parameters
qda = QuadraticDiscriminantAnalysis()

# Fit the QDA classifier on the training data
qda.fit(x_train_val_scaled, y_train_val)

# Make predictions on the test data
y_pred_qda = qda.predict(x_test_scaled)

# Calculate accuracy of QDA on the test data
accuracy_default_qda = accuracy_score(y_test, y_pred_qda)

# Calculate AUC (Area Under the ROC Curve) for QDA on the test data
y_prob_qda = qda.predict_proba(x_test_scaled)[:, 1]
auc_default_qda = roc_auc_score(y_test, y_prob_qda)
```

```python
print(f"QDA Accuracy: {accuracy_default_qda:.5f}")
print(f"QDA AUC: {auc_default_qda:.5f}")
```

QDA with default parameters has worse results than LDA for both AUC and accuracy. However, it not too bad also. Let's see if it can make a bigger difference with optimal parameters.

```python
# Create a QDA classifier
qda = QuadraticDiscriminantAnalysis()

# Define the range of values for the 'reg_param' parameter
param_grid = {'reg_param': [0.001, 0.01, 0.1, 1.0]}

# Create a scorer for ROC AUC
roc_auc_scorer = make_scorer(roc_auc_score)

# Create a GridSearchCV object
grid_search = GridSearchCV(qda, param_grid, scoring=roc_auc_scorer, cv=5)

# Fit the GridSearchCV to the training data
grid_search.fit(x_train_val_scaled, y_train_val)

# Get the best estimator and best parameter
best_qda = grid_search.best_estimator_
best_reg_param = grid_search.best_params_['reg_param']

# Evaluate the best estimator on the test data
best_qda.fit(x_train_val_scaled, y_train_val)
y_prob_best_qda = best_qda.predict_proba(x_test_scaled)[:, 1]
```

```python
accuracy_best_qda = accuracy_score(y_test, best_qda.predict(x_test_scaled))
auc_best_qda = roc_auc_score(y_test, y_prob_best_qda)
```

```python
# Print the best parameters and the corresponding AUC
print(f"Best reg param: {best_reg_param}")
```

The best reg param value, in this case, turns out to be 1.0, indicating that a higher level of regularization was preferable for ou dataset.

```python
#compare LDA and QDA results
df_models = add_model_data("QDA", accuracy_default_qda, auc_default_qda, accuracy_best_qda, auc_best_qda, df_models)
display_styled_dataframe(apply_style_to_dataframe(df_models))
```

Finally, the optimal parameters for QDA as for LDA changed only slighly the results. Indeed, LDA tends to perform a bit better for both indices (AUC and Accuracy) than QDA but the diference is small between the two methods.

### 2.3 Nonlinear methods


#### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
# Create a Random Forest classifier
rf = RandomForestClassifier()

# Fit the RF classifier on the training data
rf.fit(x_train_val_scaled, y_train_val)

# Make predictions on the test data
y_pred_rf = rf.predict(x_test_scaled)

# Calculate accuracy of RF on the test data
accuracy_default_rf = accuracy_score(y_test, y_pred_rf)

# Calculate AUC (Area Under the ROC Curve) for RF on the test data
y_prob_rf = rf.predict_proba(x_test_scaled)[:, 1]
auc_default_rf = roc_auc_score(y_test, y_prob_rf)
```

```python
print(f"RF default Accuracy: {accuracy_default_rf:.5f}")
print(f"RF default AUC: {auc_default_rf:.5f}")
display_styled_dataframe(apply_style_to_dataframe(df_models))
```

Default parameters results for Random Forest are great, the AUC and Accuracy are better than Linear methods with default parameters. Let's see if we can gain performance while finding the best parameters with a GridSearchCV.

#### Confusion Matrix

We'll work in this part with the best suitable model which is the Logistic Regression. It offers very good results and clear interpretation.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
def calculate_fnr(confusion_matrix):
    """
    Calculate the False Negative Rate (FNR) from a confusion matrix.

    Parameters:
    - confusion_matrix: numpy.ndarray
        The confusion matrix.

    Returns:
    - fnr: float
        The False Negative Rate.
    """
    TP = confusion_matrix[1, 1]  # True Positives
    FN = confusion_matrix[1, 0]  # False Negatives

    fnr = FN / (FN + TP)
    return fnr
```

```python
def calculate_fpr(confusion_matrix):
    """
    Calculate the False Positive Rate (FPR) from a confusion matrix.

    Parameters:
    - confusion_matrix: numpy.ndarray
        The confusion matrix.

    Returns:
    - fpr: float
        The False Positive Rate.
    """
    TN = confusion_matrix[0, 0]  # True Negatives
    FP = confusion_matrix[0, 1]  # False Positives

    fpr = FP / (FP + TN)
    return fpr
```

```python
y_pred_best_model = best_lr.predict(x_test_scaled)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_best_model)
conf_matrix_plot = ConfusionMatrixDisplay(conf_matrix, display_labels=best_lr.classes_)

FNR = calculate_fnr(conf_matrix)

conf_matrix_plot.plot()
```

We can see on a confusion matrix a visual representation of the classification of our model. Here on test data we can see that we have 27 False Positive (FP) and 55 False Negative (FN). In our medical context it is more important to minimize FN resulsts than FP results, why ?

In a medical context, it is more critical to minimize False Negative (FN) results than False Positive (FP) results. For instance, in the case of a widely known COVID example, having a false positive result may be merely inconvenient, as it may lead to some additional tests or precautions. However, if the model produces false negatives and fails to detect a real positive case, it could have severe consequences, potentially allowing a contagious individual to go undetected and spread the disease to others, leading to more significant health risks and potential outbreaks.

```python
print(f"\nFalse Negative Rate (FNR): {FNR:.4f}")
```

In this context, the FNR of 0.4231 means that the model fails to identify 42.31% of the true positive cases (instances of diabetes) in our dataset. In other words, it misses a significant percentage of individuals who actually have diabetes. Thus, it can be problematic, especially in a medical or diagnostic context.

Reducing the FNR is crucial in applications where failing to identify positive cases can have serious consequences. We might want to try to reduce this FNR by modifying the intercept.

**Since a false negative can have an important medical impact, propose a new threshold for the predicted
score that leads to a FNR of less that 10% (this can be done by changing manually the value of the
intercept_ in the trained classifier).**

```python
intercept_list = np.linspace(-3, 1, 100)
```

```python
FNR_list = []
FPR_list = []
accuracy_list = []

# Create a logistic regression model with the best hyperparameters
LR = LogisticRegression(C=best_C,
                        penalty="l1",
                        solver="liblinear")

LR.fit(x_train_val_scaled, y_train_val)
original_intercept = LR.intercept_

for intercept in intercept_list:
    # Set the intercept value
    LR.intercept_ = np.array([intercept])
    # Make predictions and calculate metrics
    y_pred = LR.predict(x_test_scaled)
    conf_matrix = confusion_matrix(y_test, y_pred)
    FNR_list.append(calculate_fnr(conf_matrix))
    FPR_list.append(calculate_fpr(conf_matrix))
    accuracy_list.append(accuracy_score(y_test, y_pred))
```

```python
# Find the threshold for the desired FNR
desired_fnr = 0.10
threshold_idx = np.where(np.array(FNR_list) < desired_fnr)[0][0]
new_intercept = intercept_list[threshold_idx]
threshold = -new_intercept

# Modify the intercept
LR.intercept_ = np.array([new_intercept])

# Display the new FNR
new_fnr = calculate_fnr(confusion_matrix(y_test, LR.predict(x_test_scaled)))
print(f"New False Negative Rate (FNR): {new_fnr:.4f}")

print(f"New intercept: {new_intercept:.4f}")
```

We now have succeded to make our FNR less than 10%. Let's try to understand the behaviour of FPR, FNR and accuracy while changing the intercept.

```python
plt.figure(figsize=(10, 7))
plt.plot(intercept_list, FNR_list, label='FNR')
plt.plot(intercept_list, FPR_list, label='FPR')
plt.plot(intercept_list, accuracy_list, label='Accuracy')
plt.axvline(x=new_intercept, c='r')
plt.axvline(x=original_intercept, c='b')
plt.legend()
plt.xlabel('Intercept Value')
plt.ylabel('FN AND FP RATE')
plt.grid()
```

We can observe the chosen intercept represented by the red line while the original one is represented by the blue line. Our objective is to strike a balance between minimizing the False Negative Rate (FNR) and maintaining a reasonable level of accuracy.

It's important not to excessively reduce the FNR, as doing so could lead to a significant decrease in overall accuracy and a simultaneous increase in the False Positive Rate (FPR). The key is to find an equilibrium point that optimally manages these variables, ensuring a trade-off that minimizes false negatives while still maintaining a satisfactory level of accuracy and controlling false positives.

```python
# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, LR.predict(x_test_scaled)),
                              display_labels=LR.classes_)
disp.plot()

plt.show()
```

Here we have plotted our new confusion matrix. We can see as expected with the previous curves that reducing the FNR has to come at a cost. The cost of reducing accuracy and increasing the FPR.

In summary we understood that we really have to adapt our use of the data to the context we're working on. Defining the priorities is essential. Indeed, having the best accuracy is not always the best option which can be at first hard to get.

Let's interpret the confusion matrix plotted after applying our MLP  on the test dataset:

- The class that experiences the highest number of misclassifications is "**8**", which is frequently predicted as "**1**". We'll investigate the factors contributing to this misclassification, including specific styles of "**8**" that may lead to such errors.

- In the second most common misclassification, we observe instances where "**8**" is predicted as "**7**" and instances where "**1**" is predicted as "8".

Notably, this misclassification between "**1**" and "**7**" appears more reasonable, as it can be relatively easy for both human observers and the model to confuse these two digits compared to distinguishing "**8**" from "**1**".

To address the question of the most challenging class to recognize and whether certain classes are more challenging to discriminate, we will further investigate the patterns and characteristics behind these misclassifications.

```python
misclassified_indices = np.where(y2_pred_best_mlp != y2_test)[0]

# Create a larger figure with subplots for the misclassified samples
fig, axes = plt.subplots(1, 5, figsize=(15, 3))

for i in range(5):  # Plot the first 5 misclassified samples
    axes[i].imshow(x2_test[misclassified_indices[i]].reshape(28, 28), cmap=plt.cm.gray_r, interpolation='nearest')
    axes[i].set_title(f'Predicted: {y2_pred_best_mlp[misclassified_indices[i]]}\nActual: {y2_test[misclassified_indices[i]]}')
    axes[i].axis('off')  # Remove axis labels

plt.show()
```

Here are some of the miss-classified samples. For a human, they are not too difficult to recognize but they look obvisouly different than the average digits of the dataset.

**Why are did the classifier fail to recognize them ?**

It is because the MLP is analyzing the image pixel by pixel so he can't learn very well the image like a CNN would using kernels and then sharing information with other pixels.

MLP is absolutely no the best architecture of neural networks for analyzing images as he will fail to recognize them if there is a slight change (for instance a little shift of the image)

```python
def calculate_accuracy_on_noisy_samples(model, noise_level, data, predicted_values, plot_samples=False, CNN=False):

    if CNN :
        well_classified_indices = np.where(predicted_values.argmax(axis=1) == y2_test)[0]
    else :
        well_classified_indices = np.where(predicted_values == y2_test)[0]

    if len(well_classified_indices) == 0:
        print("No well-classified samples found. Unable to calculate accuracy.")
        return None

    well_classified_sample = data[well_classified_indices[0]]

    # Create 1000 noisy samples of the well-classified sample
    if CNN :
        noisy_samples = np.random.randn(1000, 28, 28, 1) * noise_level + well_classified_sample
        # Make predictions on the noisy samples using the provided model
        y2_pred_noisy = model.predict(noisy_samples)
        # Calculate accuracy on the noisy samples
        accuracy_noisy = accuracy_score(np.ones(1000), y2_pred_noisy.argmax(axis=1))
    else :
        noisy_samples = np.random.randn(1000, 28, 28) * noise_level + well_classified_sample.reshape(28, 28)
        # Flatten the noisy samples
        noisy_samples_flat = noisy_samples.reshape(1000, -1)
        # Make predictions on the noisy samples using the provided model
        y2_pred_noisy = model.predict(noisy_samples_flat)
        # Calculate accuracy on the noisy samples
        accuracy_noisy = accuracy_score(np.ones(1000), y2_pred_noisy)

    if plot_samples and CNN is not True :

        # Select noisy samples where the classifier's predictions are incorrect
        incorrect_indices = np.where(y2_pred_noisy != 1)[0]
        incorrect_samples = noisy_samples[incorrect_indices]
        incorrect_labels = y2_pred_noisy[incorrect_indices]

        # Visualize the incorrect samples and their predicted labels
        num_samples_to_visualize = min(5, len(incorrect_samples))
        fig, axes = plt.subplots(1, num_samples_to_visualize, figsize=(12, 4))

        for i in range(num_samples_to_visualize):
            ax = axes[i]
            ax.imshow(incorrect_samples[i], cmap='gray')
            ax.set_title(f'Predicted: {incorrect_labels[i]}')
            ax.axis('off')

        plt.show()

    elif (plot_samples and CNN is True) :

        # Select noisy samples where the classifier's predictions are incorrect
        incorrect_indices = np.where(y2_pred_noisy.argmax(axis=1) != 1)[0]
        incorrect_samples = noisy_samples[incorrect_indices]
        incorrect_labels = y2_pred_noisy.argmax(axis=1)[incorrect_indices]

        # Visualize the incorrect samples and their predicted labels
        num_samples_to_visualize = min(5, len(incorrect_samples))
        fig, axes = plt.subplots(1, num_samples_to_visualize, figsize=(12, 4))

        for i in range(num_samples_to_visualize):
            ax = axes[i]
            ax.imshow(incorrect_samples[i].reshape(28,28), cmap='gray')
            ax.set_title(f'Predicted: {incorrect_labels[i]}')
            ax.axis('off')

        plt.show()


    return accuracy_noisy
```

We want to see how the model (here the MLP) is robust in recognizing images if we put him a challenge => Here adding noise to the image.

```python
noise_level = 0.20
model_to_use = best_mlp
predicted_values = y2_pred_best_mlp
accuracy_on_noisy_samples = calculate_accuracy_on_noisy_samples(model_to_use, noise_level, x2_test, predicted_values, plot_samples=False)
print(f"Accuracy on Noisy Samples: {accuracy_on_noisy_samples:.5f}")
```

No problem with a noise of 20% for the MLP, let's increase that a bit.

```python
noise_level = 0.25
accuracy_on_noisy_samples = calculate_accuracy_on_noisy_samples(model_to_use, noise_level, x2_test, predicted_values, plot_samples=True)
print(f"Accuracy on Noisy Samples: {accuracy_on_noisy_samples:.5f}")
```

At 25%, it is starting to do mistakes, and we can clearly see than the miss-classified samples are really easy to recognize and even more than the miss-classified digits without noise that we saw before.

```python
noise_level = 0.3
accuracy_on_noisy_samples = calculate_accuracy_on_noisy_samples(model_to_use, noise_level, x2_test, predicted_values, plot_samples=True)
print(f"Accuracy on Noisy Samples: {accuracy_on_noisy_samples:.5f}")
```

With 30% of accuracy it is more than struggling with only 11% accuracy. For us humans, it is still very easy to classify the numbers.

```python
noise_level = 0.4
accuracy_on_noisy_samples = calculate_accuracy_on_noisy_samples(model_to_use, noise_level, x2_test, predicted_values, plot_samples=True)
print ("x2_train shape : ", x2_train.shape)
print ("predicted values : ", predicted_values.shape)
print ("y2_test shape : ", y2_test.shape)
print(f"Accuracy on Noisy Samples: {accuracy_on_noisy_samples:.5f}")
```

At 40% noise level, the MLP model struggles to classify effectively, rendering it unsuitable for reliable image recognition.

## Bonus: Convolutional Neural network (CNN)


Let's now use the **state of the art** for image recognition which are CNN's.

We'll try so show how robust the CNN are in comparison to the MLP for image recognition.

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# Reshaping the data
x2_train_cnn = x2_train.reshape(x2_train.shape[0], 28, 28,1)
x2_test_cnn = x2_test.reshape(x2_test.shape[0], 28, 28,1)
print("(X) CNN Train shape : ", x2_train_cnn.shape)
```

```python
# Create a CNN model
model_cnn = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(10, activation='softmax')
])
```

I didn't succeed to find the best parameters of the CNN because the simple way would have been to use a KerasClassifier as we used GridSearch before but I have a problem a versions between tensorflow and keras that stop me to use it.

However, even with this architecture we'll have good results of prediction and it will clearly enough to test the robustness of the model.

```python
model_cnn.summary()
```

```python
# Compile the model
model_cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model_cnn.fit(x2_train_cnn, y2_train, epochs=10, batch_size=64)

# Evaluate the model on the test data
test_loss, test_acc = model_cnn.evaluate(x2_test_cnn, y2_test, verbose=2)
```

```python
print(f"Test Accuracy: {test_acc:.5f}")
```

We still have a really decent accuracy result of 97.6%.
